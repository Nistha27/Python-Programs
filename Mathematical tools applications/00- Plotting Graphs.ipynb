```python
#Program to plot curve connecting the points using plot and scatter function
from pylab import * #Library which comprise of numpy and matplotlib
x=[2,3,4,5,7] # x - values
y=[4,2,3,7,10] # y - values
if len(x)!=len(y):
    
    print("number of values of x and y doesnot match")
else:
    scatter(x,y) #plot points in the plane
    plot(x,y) #plot curve connecting the points
    grid()
    xlabel("x-axis")
    ylabel("y-axis")
    title("Scatter grph")
    xlim(0,10)
    ylim(0,10)
    axvline(color="red")
    axhline(color="red")
    show()
```

```python
#Program to plot exponential function
from pylab import * #Library which comprise of numpy and matplotlib
x = arange(-5,5,0.001) # x takes the values between -10 and 10 with a 
#step length of 0.001
y = exp(x) # Exponential function
plot(x,y) # plotting the points
title(" Exponential curve ") # giving a title to the graph
grid()
xlabel("x-axis")
ylabel("y-axis")
xlim(-10,10)
axvline(color="red")
axhline(color="red")
show()
```

```python
#Program to plot sin(x) and cos(x) together
from pylab import * #Library which comprise of numpy and matplotlib
x = linspace(-2*pi,2*pi,100) # x takes the values between -10 and 10 with
#a step length of 0.001
plot(x,sin(x),x,cos(x)) # plotting the points
title(" Exponential curve ") # giving a title to the graph
grid()
xlabel("x-axis")
ylabel("y-axis")
title("Scatter grph")
xlim(-2*pi,2*pi)
axvline(color="red")
axhline(color="red")
show()
```

```python
#Program to plot functions x, x^2 and X^3 for same values of x
from pylab import *
x= linspace (0 , 2 , 100 )
plot(x,x,label='linear ') # Plot of y=x a linear curve
plot(x,x**2,label='quadratic ') # Plot of y=x^2 a quadric curve
plot(x,x**3,label='cubic ') # Plot of y=x^3 a cubic curve
xlabel('x label ') # Add an x- label to the axes .
ylabel('y label ') # Add a y- label to the axes .
title(" Simple Plot ") # Add a title to the axes .
legend(loc="upper right") # Add a legend
axvline(color="red")
axhline(color="red")
ylim(0,6)
grid()
show() # to show the complete graph
```

```python
#Program to plot parametric curve circle x=r*cos(theta) and y=r sin(theta)
from pylab import *
#x=arange(-3,3,0.001)
theta=linspace(0,2*pi,1000)
x=3*cos(theta)
y=3*sin(theta)
plot(x,y)
axis("square")
title("$x=3 cos(\\theta) & y=3 sin(\\theta)$")
axvline(color="red")
axhline(color="red")
grid()
show()
```

```python
#Program to plot parametric curve cycloid x=r*(theta-sin(theta)) and
#y=r (1-cos(theta))
from pylab import *
theta=linspace(0,2*pi,1000)
x=3*(theta-sin(theta))
y=3*(1-cos(theta))
plot(x,y)
title("$x=3 (\\theta-sin(\\theta)) & y=3 (1-cos(\\theta))$")
axvline(color="red")
axhline(color="red")
grid()
show()
```

```python
# Plot cardioid r=5(1+cos theta )
from pylab import *
theta=linspace(0,2*pi,1000)
r1=5+5*cos( theta )
polar(theta,r1,'r')
title("$r=a(1+cos(\\theta))$")
show()
```

```python
#Program to pLot cardiod r=a(1+cos(theta)) take a =2
theta=linspace(0,2*pi,1000)
r=2*(1+cos(theta))
polar(theta,r)
title("cardiod: $r=a(1+cos(\\theta))$")
axvline(color="red")
show()
```

```python
#Program to pLot cardiod r=a(1+cos(theta)) &r=a(1-cos(theta))
theta=linspace(0,2*pi,1000)
r1=2*(1+cos(theta))
r2=2*(1-cos(theta))
polar(theta,r1,theta,r2)
title("cardiods: $r=a(1+cos(\\theta))$ & $r=a(1-cos(\\theta))$")
axvline(color="red")
show()
```

```python
#Program to plot four leaved rose r=2|cos(2*theta)|
from pylab import *
theta=linspace(0,2*pi,1000)
r=2*abs(cos(2*theta))
polar(theta,r)
title("$r=2|cos(2*\\theta))|$")
axvline(color="red")
show()
```

```python
#Program to plot circle r=5
from pylab import *
theta=linspace(0,2*pi,1000)
r=[5 for i in theta]
polar(theta,r,color="red")
title("$circle: r= constant $")
axvline(color="green")
show()
```

```python
#Program to plot implicit function, Cissiod: y^2(a-x)=x^3; a=2
from pylab import *
from sympy import *
x,y=symbols('x,y')
plot_implicit(Eq(y**2*(2-x),x**2*(2+x)),xlabel="",ylabel="",
 title="Strophoid:$y^2(2-x)=x^2(2+x)$")
show()
```

```python
#Program to plot implicit function, Circle: x^2 + y^2 =a^2; a=2
from pylab import *
from sympy import *
x,y=symbols('x,y')
p1=plot_implicit(Eq(x**2 + y**2,4),xlabel=False,ylabel=False,
 title='Circle : $x^2+y^2=4$ ',aspect_ratio=(2.,2.)) # a= 2
show()
```

```python
#Program to plot Cissiod - an implicit function
from sympy import *
from pylab import *
 
x,y=symbols('x y')
p2=plot_implicit(Eq ((y**2)*(3-x),x**3),xlabel=False,ylabel=False,
 title='Cissiod : $y^2 (a-x)=x^3, a> 0$ ') # a= 2
show()
```

```python
#Program to plot Lemniscate - an implicit function
from sympy import *
from pylab import *
 
x,y=symbols('x y')
p2=plot_implicit(Eq(4*(y**2),(x**2)*(4-x**2)),xlabel=False,ylabel=False,
 title='Leminscate : $a^2 y^2 = x^2 (a^2-x^2), a> 0$ ') # a= 2
show()
```
